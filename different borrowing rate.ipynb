```python
#The effect of increasing and decreasing borrowing rate by 20% on the proportion invested in production.
%pylab
import numpy as np
import math as m
from matplotlib import pylab  as plt
from mpl_toolkits.mplot3d import Axes3D


s = np.linspace(0.3,1,100)  #the range for sigma
#coefficient of the model
B = -0.5; a = 0.18; b = 0.15; R=0.8; L = 0.9; T =2; K =0.9;  S = 5; P =4; t=0


F = lambda s:m.exp((B*(a-(0.5*s**2)))*t)*(S*m.exp((a-(0.5*s**2))*t)-P*m.exp((b-(0.5*s**2))*t))

Fp = lambda s:m.exp((B*(a-(0.5*s**2)))*t)*(S*m.exp((a-(0.5*s**2))*t)*(a-(0.5*s**2))-
    (P*m.exp((b-(0.5*s**2))*t)*(b-(0.5*s**2))))+m.exp((B*(a-(0.5*s**2)))*t)*B*(a-(0.5*s**2))*(S*m.exp((a-(0.5*s**2))*t)-
    P*m.exp((b-(0.5*s**2))*t))
   #This is a function that calculate the proportion of the weath invested in the production for every value of sigma
def p(x,dev,Fun,ra1,Beta,R,ra):
    f = (dev(x)/Fun(x)) + (0.5*(1+Beta)**2*x**2)   # (pi/W)
    if ra1 <= f-(R*((1+Beta)**2)*x**2):     #   condition for borrowing rate
        fi = (f-ra1)/(R*(1+Beta)**2*x**2)
    elif ra < f-(R*((1+Beta)**2)*x**2) and f-(R*((1+Beta)**2)*x**2) < ra1:   #when r is less than r'$
        fi = 1
    else :
        fi = (f-ra)/(R*(1+Beta)**2*x**2)  # condition for deposit rate
        
    return fi

y = []
for i in s:
    y.append(p(i,Fp,F,0.1,B,R,0.05))
z = []
for i in s:
    z.append(p(i,Fp,F,0.08,B,R,0.05))  # this is for the case when borrowwing rate decreases by 200 percent
w=[]
for i in s:
    w.append(p(i,Fp,F,0.12,B,R,0.05))  # this is the case when the  borrowing rate increases by 20 percent
    # This plot show the eeffect of increasing the borrowing rate
plt.plot(s,y,)
plt.plot(s,z,)        
plt.plot(s,w,)
plt.legend()

#plt.title('t=0',loc='right')
#plt.plot(s,y, [0.413598,0.413598],[0, 1],'--',[0.533102,0.533102],[0, 1],'--',[0.6,0.6],[0, 1],'--')
plt.xlim(0.2,1.1)
plt.ylim(0,2)
plt.plot([0.54079], [1], '*')
plt.annotate('(0.54079, 1)',
             xy=(0.54079,0.9),
             xycoords='data',
             xytext = (0.5, 0.5),
             va='bottom', ha = 'left', 
             textcoords='offset points',
             arrowprops=dict());
plt.plot([0.54079], [1.2606], '*')
plt.annotate('(0.54079,1.2606)',
             xy=(0.54079,1.2606),
             xycoords='data',
             xytext = (0.5, 0.5),
             va='bottom', ha = 'left', 
             textcoords='offset points',
             arrowprops=dict());
plt.plot([0.54079], [1.60204], '*') 
plt.annotate('(0.54079, 1.60204)',
             xy=(0.54079,1.60204),
             xycoords='data',
             xytext = (0.5, 0.5),
             va='bottom', ha = 'left', 
             textcoords='offset points',
             arrowprops=dict());

plt.plot([0.771266], [0.719388], '*') 
plt.annotate('(0.771266, 0.719388)',
             xy=(0.8,0.7),
             xycoords='data',
             xytext = (0.5, 0.5),
             va='bottom', ha = 'left', 
             textcoords='offset points',
             arrowprops=dict());

plt.annotate('r1=0.1, r=0.05',
             xy=(0.48,1.8),
             xycoords='data',
             xytext = (-65, -60),
             va='bottom', ha = 'left', 
             textcoords='offset points',
             arrowprops=dict(arrowstyle = '->'));
plt.annotate('r1=0.12, r=0.65',
             xy=(0.4,1.9),
             xycoords='data',
             xytext = (-75, -30),
             va='bottom', ha = 'left', 
             textcoords='offset points',
             arrowprops=dict(arrowstyle = '->'));
plt.annotate('r1=0.08, r=0.04',
             xy=(0.52,1.9),
             xycoords='data',
             xytext = (30, -20),
             va='bottom', ha = 'left', 
             textcoords='offset points',
             arrowprops=dict(arrowstyle = '->'));


plt.ylabel('$\pi$*/w')
plt.xlabel('sigma') 
plt.savefig("hara.png")
plt.show()
```
